## Pandas, Scatter Plots

Using pandas most of the plotting functionality in pandas is contained within the `DataFrame.plot()` method. When we use this method we just have to specify the data we want to plot and the type of plot. The `kind` parameter is used to specify the type of plot that we want. We use `x` and `y` to specify the data we want on each axis. 
<br/>
In the example below we plot the `Sample_size` of those `Employed`
+ `%matplotlib inline` using this allows us to run 1 line of code to generate the plot

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
```

`DataFrame.plot()` method has a few params we can use for tweaking the scatter plot

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', 
                  title='Employed vs. Sample_size', figsize=(5,10))
```

Now the plot has updated based on the specific details we included. If we want to access the **Axes** object then we can do the following:

```python
ax = recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
ax.set_title('Employed vs. Sample_size')
```

## Plotting Histograms (Pythonic way)

Since doing the same thing we can use a loop to create the plots
+ `for r in range(0,4)`
    + Create 4 x 1 subplots

```python
# Put columns that we want plots for in a list
cols = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 
       'Men', 'Women']

fig = plt.figure(figsize=(5,12))
# Create a loop over the list 

for r in range(0,4):
    ax = fig.add_subplot(4,1,r+1)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30)
```

## Pandas - Scatter Matrix Plot

If we want to visulize potential relationships between columns and histograms to visualize the distributions of individual columns we can use a scatter matrix plot.

+ scatter matrix plot combines historgram and scatter plots in a grid
+ frequently used in the exploratory data analysis
+ in pandas we can use the `scatter_matrix()` to generate the plots for us
    + this function is part of the `pandas.plotting` module and needs to be imported separately
    + if you want just 2 columns then you select the columns and pass the resulting DataFrame into the `scatter_matrix()` function.

```python
# example
# have to import scatter_matrix module from pandas.plotting

import pandas.plotting
from pandas.plotting import scatter_matrix


scatter_matrix(recent_grads[['Women', 'Men']], figsize=(10,10));
```

```python
# Creates 2x2 scatter matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']]);
```

```python
# Create 3x3 scatter matrix
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10));
```

## Pandas - Bar Plots

For bar plots we have to specify many aspects of the bar plot ourselves:
+ locations
+ labels
+ lengths
+ widths, etc.

If we use pandas to create bar plots then we only have to specify the columns that we want to be represented in bars and the labels for each bar.
<br/><br/>
In the code below returns a bar plot of the first 5 values in the `Women` column:

```python
recent_grads[:5]['Women'].plot(kind='bar');
```

If we want to specify then we can use the `DataFrame.plot.bar()` method

```python
recent_grads[:5].plot.bar(x='Major', y='Women');
```
