```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv', encoding="ISO-8859-1")
```

```python
# Review the column names
star_wars.columns
```

```python
# Remove rows where RespondentID is NaN
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

```python
star_wars.head()
```

```python
# Checking that null rows have been removed
star_wars['RespondentID'].notnull().value_counts()
```

# Cleaning and Mapping Yes/No Columns

Currently these columns contain `Yes` and `No` responses. We will change them to Boolean's which have the values `True`, `False`, and `NaN`. Booleans are easier to work with because we can select the rows that are `True` or `False` without having to do a string comparison.

+ we will use `pandas.Series.map()` to perform the conversion

```python
# Look at the unique responses in these two columns
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
# use a dictionary to define a mapping from each value in the series to a new value
yes_no = {'Yes': True, 'No': False}

# call the map() function to perform the mapping
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
star_wars.head()
```

# Cleaning and Mapping Checkbox Columns

The six columns represent a single checkbox question. If a respondent checked off a series of boxes showing they have watched the following Star Wars movies.
+ If the box contains the title of the movie the respondent has watched it, otherwise if it contains `NaN` they haven't watched the movie or did not answer the question. The prior is assumed. 

+ Approach
    + Convert each of these columns to a Boolean
    + Rename the column something intuitive
    + Convert the values using mapping including movie title and NaN

```python
# Select the column names
star_wars.columns[3:9]
```

```python
# create movie dictionary
import numpy as np

movies = {"Star Wars: Episode I  The Phantom Menace" : True, 
          np.NaN: False,
          "Star Wars: Episode II  Attack of the Clones": True,
          "Star Wars: Episode III  Revenge of the Sith": True,
          "Star Wars: Episode IV  A New Hope": True,
          "Star Wars: Episode V The Empire Strikes Back": True,
          "Star Wars: Episode VI Return of the Jedi": True}
```

```python
# call the map() function to perform the mapping
# need to iterate over the columns from indexs 3 to 9
for cols in star_wars.columns[3:9]:
    star_wars[cols] = star_wars[cols].map(movies)
```

```python
# rename the columns and assign back to dataframe
star_wars = star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply." : 'seen_1',
                                     'Unnamed: 4' : 'seen_2',
                                     'Unnamed: 5' : 'seen_3',
                                     'Unnamed: 6' : 'seen_4',
                                     'Unnamed: 7' : 'seen_5', 
                                     'Unnamed: 8' : 'seen_6'})
```

```python
star_wars.head()
```

# Cleaning the Ranking Columns

The next 6 columns respondents were asked to rank the Star Wars movies in order of least favorite to most favorite (1 being most favorite)

+ Approach:
    + Convert each column to a numeric conversion using `pandas.DataFrame.astype()`

```python
# converting to numeric
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
# check that conversion of cols
star_wars.head(2)
```

```python
# rename ranking columns
star_wars = star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film." : 'ranking_1',
                                     'Unnamed: 10' : 'ranking_2',
                                     'Unnamed: 11' : 'ranking_3',
                                     'Unnamed: 12' : 'ranking_4',
                                     'Unnamed: 13' : 'ranking_5', 
                                     'Unnamed: 14' : 'ranking_6'})
```

```python
# check cols
star_wars.head(0)
```

# Find the Highest-Ranked Movie

```python
# getting the mean of the ranking columns
star_wars[star_wars.columns[9:15]].mean()
```

```python
# plot the ranking means
import matplotlib.pyplot as plt
%matplotlib inline
plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
plt.xlabel('ranking')
plt.ylabel('Avg. ranking')
```

So far, we've cleaned up the data, renamed several columns, and computed the average ranking of each movie. As I suspected, it looks like the "original" movies are rated much more highly than the newer ones.


# Most viewed movie

It appears that the original movies were seen by more respondents than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular.

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
plt.xlabel('Movies watched')
plt.ylabel('Sum of Views')
```

# Exploring the Data by Binary Segments



```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
# Rankings
plt.bar(range(6), males[males.columns[9:15]].mean(), label='Males')
plt.bar(range(6), females[females.columns[9:15]].mean(), label='Females', color='red')
plt.legend(loc='best')
plt.xlabel('Assigned Rankings')
plt.ylabel('Avg. Rankings')
```

```python
# Most watched
plt.bar(range(6), males[males.columns[3:9]].sum(), label='Males')
plt.bar(range(6), females[females.columns[3:9]].sum(), label='Females', color='red')
plt.legend(loc='best')
plt.xlabel('Movies watched')
plt.ylabel('Sum of Views')
```

more males watches episodes 1-3, but males liked them far less than females did.
